```python
def hypothesis(X,theta):
    return np.dot(X,theta)
```

```python
def error(X,y,theta):
    e=0.0
    m=X.shape[0]
    h=hypothesis(X,theta)
    e=np.sum((y-h)**2)
    return e/m
```

```python
def gradient(X,y,theta):
    h=hypothesis(X,theta)
    grad=np.dot(X.T,(h-y))
    m=X.shape[0]
    return grad/m
```

```python
def gradient_descent(X,y,lr=0.1,max=300):
    n=X.shape[1]
    theta=np.zeros((n,))
    error_list=[]
    for i in range(max):
        e=error(X,y,theta)
        error_list.append(e)
        grad=gradient(X,y,theta)
        theta=theta-lr*grad
    return theta,error_list    
        
```

```python
theta,error_list=gradient_descent(X,y)
```

```python
plt.plot(error_list)
plt.show()
```

```python
predict=hypothesis(X,theta)
```

```python
def R2_score(y,h):
    num=np.sum((y-h)**2)
    deno=np.sum((y-y.mean())**2)
    score=1-(num/deno)
    return score*100
```

```python
print(R2_score(y,predict))
```
